###### Assigning categories to Age and Household income for better analysis for strong predictor.

**Assigning low, medium and high categories to the salery ranges.**

```python
df['Household Income'].value_counts()
```

```python
df['Household Income'].replace('Prefer not to answer','$175,000-$199,999', inplace=True)
```

```python
df['Household Income'].value_counts()
```

```python
# Mapping dictionary for income categories
income_mapping = {
    '$0-$9,999': 'Low',
    '$10,000-$24,999': 'Low',
    '$25,000-$49,999': 'Medium',
    '$50,000-$74,999': 'Medium',
    '$75,000-$99,999': 'Medium',
    '$100,000-$124,999': 'High',
    '$125,000-$149,999': 'High',
    '$150,000-$174,999': 'High',
    '$175,000-$199,999': 'High',
    '$200,000+': 'High'
}
```

```python
# Function to assign income categories
def assign_income_category(income_range):
    return income_mapping.get(income_range) 
```

```python
df['HHI_Category'] = df['Household Income'].apply(assign_income_category)
```

```python
df['Household Income'].head()
```

```python
df.head()
```

```python
df['HHI_Category'].value_counts()
```

```python
dummy_hhi_cat = pd.get_dummies(df['HHI_Category'])
dummy_hhi_cat.head()
```

```python
dummy_hhi_cat.rename(columns={'High':'hhi-high', 'Low':'hhi-low','Medium':'hhi-medium'}, inplace=True)
dummy_hhi_cat.head()
```

```python
# concatinating 
df = pd.concat([df,dummy_hhi_cat], axis=1)
# dropping original column aspiration
df.drop('HHI_Category', axis =1, inplace= True)
```

```python
df.head()
```

**Writing a function to assign categories names to the ranges of age.**
1. 18-29/30-44 years - I have called this group young adults
2. 45-60 years - I have called this group adults
3. Greater than 60 years - I have called this group seniors

```python
df['Age'].value_counts()
```

```python
# Function to assing categories 
def assign_age_category(age_range):
    if age_range in ['30-44', '18-29']:
        return 'Young Adults'
    elif age_range == '45-60':
        return 'Adults'
    elif age_range == '> 60':
        return 'Seniors'
    else:
        return 'Other'  # Handle other cases if needed
```

```python
# Apply the function to create the 'Age_Category' column
df['Age_Category'] = df['Age'].apply(assign_age_category)
```

```python
df['Age_Category'].head()
```

```python
df.head()
```

```python
df['Age_Category'].value_counts()
```

```python
dummy_age_cat = pd.get_dummies(df['Age_Category'])
dummy_age_cat.head()
```

```python
dummy_age_cat.rename(columns={'Adults':'age-adults', 'Seniors':'age-seniors','Young Adults':'age-young-adults'}, inplace=True)
dummy_age_cat.head()
```

```python
# concatinating 
df = pd.concat([df,dummy_age_cat], axis=1)
# dropping original column aspiration
df.drop('Age_Category', axis =1, inplace= True)
```
